# Problem 1 - PCA

1. Create the data set:
- Import the MNIST data set (from Keras)
- Select the images corresponding to single digit (e.g. 7)
- Run the PCA analysis on images

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

digit = 7

x_digit= x_train[y_train == digit]


#flatten the image inot a vector using .shape
X = x_digit.reshape(x_digit.shape[0], -1)

pca = decomposition.PCA(n_components=784)
pca.fit(X)
```

2. Visualize the scree plot

```python
#shows variance ratio vs number of principle components

# Explained variance ratio: fraction of dataset variance explained by each PC
explained_var = pca.explained_variance_ratio_

# Make a scree plot
plt.figure(figsize=(6,4))

# Plot line with markers
plt.plot(np.arange(1, len(explained_var)+1), explained_var, marker='o')

# Add axis labels and title
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot (Variance Explained by Each Component)")

# Add a grid for easier reading
plt.grid(True)

plt.show()
```

3. Visualize the PCA components and loadings distribution (the first two components)

```python
# 1. Visualize the first two PCA components as images (eigen-digits)
fig, axes = plt.subplots(1, 2, figsize=(6,3))

for i, ax in enumerate(axes):
    # Each component is length 784, reshape back to 28x28
    ax.imshow(pca.components_[i].reshape(28,28), cmap='gray')
    ax.set_title(f'PC {i+1}')
    ax.axis('off')

plt.suptitle("First Two PCA Components (Eigen-digits)")
plt.show()


# 2. Visualize loadings (scores of images in PC1-PC2 space)
X_pca = pca.transform(X)  # project original images into PCA space

plt.figure(figsize=(6,6))
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.5, s=10)
plt.xlabel("PC1 Score")
plt.ylabel("PC2 Score")
plt.title("Distribution of Digit 7 in PCA Space (PC1 vs PC2)")
plt.grid(True)
plt.show()
```

4. Generative model:
- run the PCA analysis with 2 components
- sample the loading space with the rectangular grid
- generate the images from the loading space (i.e. create latent representation)
- do you observe any trends in images across the loading space?

```python
pca_2 = decomposition.PCA(n_components=2)
X_pca2 = pca_2.fit_transform(X)

grid_x = np.linspace(-10, 10, 15)
grid_y = np.linspace(-10, 10, 15)

fig, axes = plt.subplots(len(grid_y), len(grid_x), figsize=(10,10))


for i, y in enumerate(grid_y):
    for j, x in enumerate(grid_x):
        # Latent point = combination of PC1 & PC2
        z = np.array([x, y])

        # Reconstruct back to image space
        img = pca_2.inverse_transform(z)

        # Reshape to 28x28
        axes[i, j].imshow(img.reshape(28,28), cmap="gray")
        axes[i, j].axis('off')

plt.suptitle("Generative PCA model (digit 7 across PC1 & PC2 space)")
plt.show()
```

# Problem 2 - LDA

7. Here, we will run the LDA on the MNIST data set.
- Identify (any) 3 digits that you would like to explore and assign as classes 1, 2, and 3
- Designate all other digits as class 0
- Run the LDA analysis
- If the analysis takes too long (>5 min), decimate the data set as [::10]

```python
#Linear Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

#flatten images
X = x_train.reshape(x_train.shape[0], -1)
y = y_train.copy()

#define classes for the numbers to explore
class_1, class_2, class_3 = 3, 4, 8

#keep only these three digits
mask = np.isin(y, [class_1, class_2, class_3])
X = X[mask]
y = y[mask]

#decimate sampling
X = X[::10]
y = y[::10]

#run LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

#finds directions that maximize class seperation
```

8. Visualize the LDA results in the 2D LDA feature space. Use the original digits as labels

```python
plt.figure(figsize=(7,7))
colors = {class_1:'red', class_2:'blue', class_3:'green'}

for digit, color in colors.items():
    plt.scatter(X_lda[y==digit,0], X_lda[y==digit,1],
                s=10, alpha=0.6, color=color, label=f'Digit {digit}')

plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.title("LDA projection of digits 3, 4, and 8")
plt.legend()
plt.grid(True)
plt.show()

#very good seperation between the 3 digits
```

9. Run the analysis when the selected classes are "0", "1", and "8". Where will be the "6" and "9" in the LDA feature space

```python
#only train the LDA on 0, 1, 8

#flatten
X = x_train.reshape(x_train.shape[0], -1)
y = y_train.copy()

#training digits
class_1, class_2, class_3 = 0, 1, 8

mask = np.isin(y, [class_1, class_2, class_3])
X_train_sub = X[mask]
y_train_sub = y[mask]


X_train_sub = X_train_sub[::10]
y_train_sub = y_train_sub[::10]

#fit LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_train_sub, y_train_sub)

#Visualize LDA space for trained classes
plt.figure(figsize=(7,7))
colors = {0:'red', 1:'blue', 8:'green'}

for digit, color in colors.items():
    plt.scatter(X_lda[y_train_sub==digit,0],
                X_lda[y_train_sub==digit,1],
                s=10, alpha=0.6, color=color, label=f'Digit {digit}')

plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.title("LDA Projection for digits 0, 1, and 8")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#project 6 and 9 onto the same space with this limited training
mask_69 = np.isin(y, [6, 9])
X_69 = X[mask_69]
y_69 = y[mask_69]

# Project them using the trained LDA model (no refit!)
X_69_proj = lda.transform(X_69)

# Overlay them on the same plot
plt.figure(figsize=(7,7))

# Plot training digits first
for digit, color in colors.items():
    plt.scatter(X_lda[y_train_sub==digit,0],
                X_lda[y_train_sub==digit,1],
                s=10, alpha=0.4, color=color, label=f'Train Digit {digit}')

# Overlay 6 and 9
plt.scatter(X_69_proj[y_69==6,0], X_69_proj[y_69==6,1],
            s=10, color='orange', label='Digit 6', alpha=0.7)
plt.scatter(X_69_proj[y_69==9,0], X_69_proj[y_69==9,1],
            s=10, color='purple', label='Digit 9', alpha=0.7)

plt.xlabel("LDA Component 1")
plt.ylabel("LDA Component 2")
plt.title("LDA trained on 0, 1, 8 — projecting 6 and 9")
plt.legend()
plt.grid(True)
plt.show()
```

Answer 9: 8 and 0 have simi8lar shapes, "9" falls as a combination of 8 and 1. "6" falls closer to the 0 and 8 cluster

Important considerations for subsequent analysis.
 - For "real" analysis, make sure that before any subsequent analysis the data is properly curated. For EELS this is zero loss peak subtraction and log-log deconvolution. This will affect analysis very strongly (e.g. zero-loss peak will dominate everything if not accounted for).
 - You can also consider downsampling the data (just average pool).
 - Finally, always keep track on whether you normalize data and how (all spectra 0 to 1, or all normalized to total maximum)
 - And keep track on whether average is subtracted.
 - Normalization and average subtraction will affect the analysis results. In our experience, there is no right way or wrong way for doing it - different preprocessing highlights dissimilar aspects of the physics of the system. However, it is important to keep track of these.  

```python
X_vec = hdata.copy()
d1, d2, d3 = X_vec.shape
X_vec = X_vec.reshape(d1*d2, d3)
X_vec = (X_vec - np.min(X_vec)) / np.ptp(X_vec)
```

Here we run the PCA analysis on EELS to visualzie the results

```python
# Select number of components
nc = 12

# Select decomposition type ('PCA' or 'NMF')
decomposition_type = 'PCA'

# Run decomposition
if decomposition_type == 'NMF':
    clf = decomposition.NMF(n_components=nc, random_state=42)
elif decomposition_type == 'PCA':
    clf = decomposition.PCA(n_components=nc, random_state=42)
else:
    raise NotImplementedError('Available methods: "PCA", "NMF"')
X_vec_t = clf.fit_transform(X_vec)
components = clf.components_
components = components.reshape(nc, d3)
print(components.shape, X_vec_t.shape)
```
